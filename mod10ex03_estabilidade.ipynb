1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
sns.set_theme(style='whitegrid')
ax = sns.countplot(x='data_ref',  hue = 'posse_de_veiculo', data=df_raw, palette= 'pastel')

tick_labs = df_raw['data_ref'].unique()

ticks = ax.set_xticks(list(range(df_raw['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)
title = ax.set_title('Posse de Veículo')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'posse_de_veiculo',
                   data=df_raw, dodge=True, ci = 95, palette= 'pastel')

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

tick_labs = df_raw['data_ref'].unique()

ticks = ax.set_xticks(list(range(df_raw['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)
title = ax.set_title('Posse de Veículo')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'posse_de_imovel',
                   data=df_raw, dodge=True, ci = 95, palette= 'pastel')

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

tick_labs = df_raw['data_ref'].unique()

ticks = ax.set_xticks(list(range(df_raw['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)
title = ax.set_title('Posse de Imóvel')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'tipo_renda',
                   data=df_raw, dodge=True, ci = 95, palette= 'pastel')

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

tick_labs = df_raw['data_ref'].unique()

ticks = ax.set_xticks(list(range(df_raw['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)
title = ax.set_title('Tipo de Renda')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'educacao',
                   data=df_raw, dodge=True, ci = 95, palette= 'pastel')

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

tick_labs = df_raw['data_ref'].unique()

ticks = ax.set_xticks(list(range(df_raw['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)
title = ax.set_title('Educação')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````
