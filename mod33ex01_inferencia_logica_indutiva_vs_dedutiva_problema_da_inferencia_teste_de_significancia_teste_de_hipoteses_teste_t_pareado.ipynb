## 1) Diferena de peso entre machos e fmeas

Na base de pinguins, podemos ver que h diferena de peso entre machos e fmeas. Mas ser que  uma concluso extrapolvel a toda a populao de pinguins? Assumindo que essa amostra  aleatria e representativa da populao de pinguins, realize um teste T com 5% de significncia para verificar se podemos induzir esta concluso para a populao toda.

```python
desvpad = pg.groupby('sex')['body_mass_g'].var().mean()**.5
Mdia_A = pg.groupby('sex')['body_mass_g'].mean()['Male']
Mdia_B = pg.groupby('sex')['body_mass_g'].mean()['Female']

diferena = Mdia_A - Mdia_B      # Numerador da frmula
erro_padrao = desvpad*((2/10)**.5) # varincia amostral*raiz(2/n)
t = diferena/erro_padrao          # Estatstica do teste
RC = st.t.ppf(.95, 18)             # Regio crtica: sob h0, t est na RC 5% das vezes
p_value=1-st.t.cdf(t, 18)          # p-value: a menor significncia para a qual rejeitamos o teste

print(f'diferena: {diferena:.2f}')
print(f'desvio padro: {desvpad:.2f}')
print(f'erro padro: {erro_padrao:.2f}')
print(f'estatstica do teste: {t:.2f}')
print(f'regio crtica: {RC:.2f} ou mais')
print(f'p-value: {p_value:.5f} ou mais')
```

**Anlise** 

Dado que o valor de p obtido  menor que 5%, podemos concluir que h uma diferena significativa de peso entre machos e fmeas.
