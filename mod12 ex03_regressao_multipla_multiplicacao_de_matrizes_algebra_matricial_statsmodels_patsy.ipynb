````python
import numpy as np
import pandas as pd

import seaborn as sns
from seaborn import load_dataset
import matplotlib.pyplot as plt

import patsy

import statsmodels.api as sm
import statsmodels.formula.api as smf
````

````python
tips = load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

### I. Modelo no valor da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo em ```tip```, explicada por ```sex, smoker, diner e net_bill```.  
2. Remova as variáveis não significantes.  
3. observe o gráfico de resíduos em função de ```net_bill```  
4. teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.

````python
#1)
y, X = patsy.dmatrices('tip ~ sex + smoker + time + net_bill + 1', tips)

X
````

````python
reg_a = sm.OLS(y,X).fit()

reg_a.summary()
````

````python
#2)
reg_b = smf.ols('tip ~ net_bill + 1', tips).fit()

reg_b.summary()
````

````python
#3)
tips['res_b'] = reg_b.resid
sns.scatterplot(data=tips, x='net_bill',y='res_b')
plt.axhline(y=0, color='r', linestyle='--')
````

````python
#4)
reg_c = smf.ols('tip ~ np.log(net_bill) + 1', tips).fit()

reg_c.summary()
````

````python
tips['res_c'] = reg_c.resid

sns.scatterplot(x = 'net_bill', y = 'res_c', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
# Polinômio (função de segundo grau)
reg_d = smf.ols('tip ~ net_bill + I(net_bill**2) + 1', tips).fit()

reg_d.summary()
````

````python
tips['res_d'] = reg_d.resid

sns.scatterplot(x = 'net_bill', y = 'res_d', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

- O melhor modelo é o primeiro. A regressão com log apresenta R² inferior a do primeiro modelo e na regressão com um polinômio de segundo grau o termo quadrático da equação possui P-value > 0.05, ou seja, o termo não é estatisticamente relevante.

### II. Modelo no valor do percentual da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo no log de ```tip```, explicado por ```sex, smoker, dinner e net_bill```.
2. Remova as variáveis não significantes.
3. Observe o gráfico de resíduos em função de ```net_bill```
4. Teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.
5. Do modelo final deste item, calcule o $R^2$ na escala de ```tip``` (sem o log). Compare com o modelo do item 1. Qual tem melhor coeficiente de determinação?

````python
#1)
y, X = patsy.dmatrices('np.log(tip) ~ sex + smoker + time + net_bill + 1', tips)
reg_e = sm.OLS(y, X).fit()
reg_e.summary()
````

````python
# 2. 'sex', 'smoker' e 'time' possuem P-value > 0.05 podemos remover do modelo
y, X = patsy.dmatrices('np.log(tip) ~ net_bill + 1', tips)
reg_f = sm.OLS(y, X).fit()
reg_f.summary()
````

````python
# 3)
tips['res_f'] = reg_f.resid

sns.scatterplot(x = 'net_bill', y = 'res_f', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
# 4)
reg_g = smf.ols('np.log(tip) ~ np.log(net_bill) + 1', tips).fit()

reg_g.summary()
````

````python
tips['res_g'] = reg_g.resid

sns.scatterplot(x = 'net_bill', y = 'res_g', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
# Usando Polinômio de segundo grau
reg_h = smf.ols('np.log(tip) ~ net_bill + I(net_bill**2) + 1', tips).fit()

reg_h.summary()
````

````python
tips['res_h'] = reg_h.resid

sns.scatterplot(x = 'net_bill', y = 'res_h', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
# Usando polinômio de terceiro grau

reg_i = smf.ols('np.log(tip) ~ net_bill + I(net_bill**2) + I(net_bill**3) + 1', tips).fit()

reg_i.summary()
````

````python
tips['res_i'] = reg_i.resid

sns.scatterplot(x = 'net_bill', y = 'res_i', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

- O modelo com polinômio de segundo grau possui um R² de 0.338 (maior os demais modelos) e os termos dos modelo possuem P-value < 0.05 o que não ocorre no polinômio de terceiro grau.

````python
#5) R²
tips['pred_h'] = reg_h.fittedvalues

r = tips[['pred_h', 'tip']].corr().iloc[0,1]

r**2
````

- Entre os 2 modelos que escolhemos o melhor valor de R² é o da regressão _reg_b_ (0.333 > 0.319).

````python
# Removendo dados faltantes
renda_a = renda.dropna()
````

````python
# 2.
tips['res_renda_a'] = reg_renda_a.resid

sns.scatterplot(x = 'net_bill', y = 'res_renda_a', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

Optei por apenas fazer uma regressão log e uma de segunda ordem em relação a variável explicativa para observar se o R² aumenta.

````python
reg_renda_b = smf.ols('renda ~ np.log(tempo_emprego) + 1', renda_a).fit()

reg_renda_b.summary()
````

````python
reg_renda_c = smf.ols('renda ~ tempo_emprego + I(tempo_emprego**2) + 1', renda_a).fit()

reg_renda_c.summary()
````

- Com o uso do polinômio de segundo grau aumentamos o R²
