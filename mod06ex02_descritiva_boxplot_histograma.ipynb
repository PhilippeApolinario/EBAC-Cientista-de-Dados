####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Usando pyplot do matplotlib
plt.boxplot(x = 'IDADEMAE', data = sinasc_ro)
plt.ylabel('IDADEMAE')
```

```python
# Usando o seaborn
sns.boxplot(y='IDADEMAE', data = sinasc_ro, color='green').set_ylabel('Idade da Mãe', fontsize=12)
```

####  2) Faça um boxplot do peso do bebê

```python
# Usando pyplot do matplotlib
plt.boxplot(x='PESO', data = sinasc_ro)
plt.ylabel('PESO')
```

```python
# Usando seaborn
sns.boxplot(y='PESO', data= sinasc_ro).set_ylabel('Peso (g)', fontsize=12)
```

####  3) Faça um histograma da idade da mãe

```python
# Usando o pyplot do matplolib
plt.hist(x ='IDADEMAE', data = sinasc_ro, bins = 20 )
plt.xlabel('IDADEMAE')
```

```python
# Usando o displot do seaborn
sns.displot(data = sinasc_ro
            , x = 'IDADEMAE'
            , bins = 42
            , alpha = .5
            , kde = True
            , element = 'step'
           ).set(xlabel='Idade da Mãe')
```

#### 4) Faça um histograma do peso do bebê

```python
# Usando pyplot do matplolib
plt.hist(x='PESO', data = sinasc_ro, bins=20 )
plt.xlabel('PESO')
```

```python
# Usando hisplot do seaborn
sns.histplot(x = 'PESO'
             , data = sinasc_ro
             , bins = 20
             , alpha = .5
             , kde = True
             , element = 'step'
            )
```
