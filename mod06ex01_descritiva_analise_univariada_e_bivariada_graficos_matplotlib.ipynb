####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
#Plot
sns.countplot(data=df_sinasc_ro , x=df_sinasc_ro['APGAR1'], alpha = 0.7, edgecolor = 'black')
plt.title("APGAR1")
ax= plt.gca()
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.4, p.get_height()), ha = 'center', color = 'black', size = 10)
```

```python
#Plot
sns.countplot(data=df_sinasc_ro , x=df_sinasc_ro['APGAR5'], alpha = 0.7, edgecolor = 'black' )
plt.title("APGAR5")
ax= plt.gca()
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.4, p.get_height()), ha = 'center', color = 'black', size = 10)
```

```python
#Plot
sns.countplot(data=df_sinasc_ro , x=df_sinasc_ro['SEXO'],alpha = 0.7, edgecolor = 'black'  )
plt.title("SEXO")
ax= plt.gca()
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', 
                (p.get_x()+0.4, p.get_height()), 
                ha = 'center', 
                va = 'top', 
                color = 'white', 
                size = 12)
```

```python
#Plot
ax = sns.catplot(data=df_sinasc_ro , x='GESTACAO',kind='count', alpha = 0.7, edgecolor = 'black')
ax.set_xticklabels(rotation=45)
plt.title("GESTAÇÃO")
ax= plt.gca()
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.4, p.get_height()), ha = 'center', color = 'black', size = 10)
```

```python
#Plot
ax = sns.countplot(data=df_sinasc_ro , x=df_sinasc_ro['ESTCIVMAE'], alpha = 0.7, edgecolor = 'black')
plt.title("ESTADO CIVIL DA MÃE")
ax.set_xticklabels(['Solteira', 'Casada', 'Viuva', 'Divorciada', 'União Consensual', 'Ignorado'],
                            rotation=30, fontsize='small') 
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.4, p.get_height()), ha = 'center', color = 'black', size = 10)
 
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
#Plot

#Tamanho da figura
plt.figure(figsize = (20,30))

#Lista de Colunas
column_list = df_sinasc_ro.columns.values[[5, 11, 14, 17]]

#Contador
A = 0

#Loop
for i in column_list:
    A += 1
    plt.subplot(5, 2, A)
    ax = sns.countplot(data = df_sinasc_ro, x = i, alpha = 0.7, edgecolor = 'black' )
    plt.title(i, fontsize = 15)
    for p in ax.patches:
        ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.4, p.get_height()), ha = 'center', color = 'black', size = 12)
    if A >= 5:
        plt.xticks(rotation = 45)
        
#Layout
plt.tight_layout(h_pad = 2)
```
